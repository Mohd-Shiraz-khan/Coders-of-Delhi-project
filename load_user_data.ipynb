```python
import json
```

```python
# Load Json File
def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data
```

```python
data = load_data("data.json")
```

```python
# Load Users details
def display_userdata(data):
    for user in data['users']:
        print(f"{user['name']} is friends with: {user['friends']} - Liked_Pages: {user['liked_pages']}")
    for page in data['pages']:
        print(f"{page['id']} - {page['name']}")
display_userdata(data)
```
